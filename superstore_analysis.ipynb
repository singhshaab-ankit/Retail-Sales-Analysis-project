```python
import pandas as pd

#Load the dataset
file_path="D:/data_analyst/AnalyticsProjects/projects/RetailSalesAnalysis/Superstore_working.csv"

df_store=pd.read_csv(file_path,encoding='latin1')
df_store.info()
```

```python
# Show first few rows
df_store.head(5)
```

```python
#to check all columns name
df_store.columns
```

```python
#Data Cleaning & Preparation Process
# Convert 'Order Date' and 'Ship Date' to datetime

df_store['Order Date'] = pd.to_datetime(df_store['Order Date'])
df_store['Ship Date'] = pd.to_datetime(df_store['Ship Date'])

# Check for duplicates
duplicate_count = df_store.duplicated().sum()

# Drop duplicates if any
df_store = df_store.drop_duplicates()

# Check for null values
null_counts = df_store.isnull().sum()

# Standardize column names (lowercase, underscores)
df_store.columns=df_store.columns.str.strip().str.lower().str.replace(' ','_')

duplicate_count, null_counts, df_store.columns.tolist()
```
